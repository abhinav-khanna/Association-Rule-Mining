```python
import random

items = {}

num_items = 100
num_trans = 1000

for i in range(1,num_items):
    ls = []
    for j in range(i):
        tid = random.randint(1,num_trans)
        ls.append(tid)
    
    items[i] = ls

print(items)
```

```python
transactions = {}

for i in range(1,num_trans+1):
    transactions[i] = []

for item in items.keys():
    ls = items[item]
    for i in ls:
        transactions[i].append(item)

print(transactions)
```

```python
f = open('generated_dataset.txt','a')
for t in transactions.keys():
    ls = transactions[t]
    for item in ls:
        f.write(str(item) + ' ')
    
    f.write('\n')
```
